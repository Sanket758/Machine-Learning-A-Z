#Importing the libraries

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Importing the dataset

```
# Read the csv contatining data
dataset = pd.read_csv('/content/Mall_Customers.csv') 
X = dataset.iloc[:,3:5].values # Getting Feature columns
```

# Training the KMeans Clustering Model

```
from sklearn.cluster import KMeans
wcss = [] #within cluster sum of squares
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```
plt.plot(range(1,11), wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('wcss')
plt.show();
```

# Training using optimal K - taken from elbow method

```
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
y_hat = kmeans.fit_predict(X)  
```

```
y_hat
```

# Visualizing clusters

```
plt.figure(figsize=(12,8));
plt.scatter(X[y_hat==0, 0], X[y_hat==0, 1], s=100, c='red', label='cluster1')
plt.scatter(X[y_hat==1, 0], X[y_hat==1, 1], s=100, c='blue', label='cluster2')
plt.scatter(X[y_hat==2, 0], X[y_hat==2, 1], s=100, c='green', label='cluster3')
plt.scatter(X[y_hat==3, 0], X[y_hat==3, 1], s=100, c='cyan', label='cluster4')
plt.scatter(X[y_hat==4, 0], X[y_hat==4, 1], s=100, c='yellow', label='cluster5')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=200,
            c='black', marker='+', label='centroids')
plt.title('Cluster of customers')
plt.xlabel('Annual Income')
plt.ylabel('Spending score (1-100)')
plt.legend()
plt.show();
```
